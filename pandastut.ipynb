```python
# Maps
price_mean = melbourne_home_data.Price.mean()
melbourne_home_data.Price.map(lambda p: p - price_mean)
```

```python
def remean_prices(row):
    row.Price = row.Price - price_mean
    return row
# Apply: used for transforming a whole DataFrame,
# by calling a custom method on each row
melbourne_home_data.apply(remean_prices, axis='columns')
```
